```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Load dataset
file_path = "/content/BTCUSD_5Y_1MONTH_FROM_PERPLEXITY.csv"
df = pd.read_csv(file_path)

# Keep only Close price
data = df[['close']].values

# Scaling to 0–1 range
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Create dataset with timestep (60 days lookback)
x = []
y = []
lookback = 60

for i in range(lookback, len(scaled_data)):
    x.append(scaled_data[i - lookback:i, 0])
    y.append(scaled_data[i, 0])

x = np.array(x)
y = np.array(y)

# Reshape to LSTM input [samples, timesteps, features]
x = np.reshape(x, (x.shape[0], x.shape[1], 1))

# Build RNN model
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(x.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(64))
model.add(Dropout(0.2))
model.add(Dense(1))  # Output layer

model.compile(optimizer='adam', loss='mean_squared_error')

# Train
history = model.fit(x, y, epochs=30, batch_size=32, verbose=1)

# Predict on same dataset (for visualization)
pred = model.predict(x)
predicted_prices = scaler.inverse_transform(pred)
actual_prices = scaler.inverse_transform(y.reshape(-1, 1))

# Plot
plt.figure(figsize=(10, 6))
plt.plot(actual_prices, label="Actual Price")
plt.plot(predicted_prices, label="Predicted Price")
plt.title("BTC Closing Price Prediction with RNN (LSTM)")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Predict next day price
last_60 = scaled_data[-lookback:]
last_60 = np.reshape(last_60, (1, lookback, 1))

next_scaled = model.predict(last_60)
next_price = scaler.inverse_transform(next_scaled)

print("Next predicted BTC closing price:", next_price[0][0])
```
